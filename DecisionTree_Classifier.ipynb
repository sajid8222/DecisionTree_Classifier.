```python
import numpy as np # linear algebra
import pandas as pd # data processing, 
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('../input/multiple-linear-regression-for-salaries/mldata.csv')
```

```python
df.head(30)
```

```python
df['gender'] = df['gender'].replace("Male",1)
df['gender'] = df['gender'].replace("Female",0)
df.tail(10)
```

```python
# Selection of input and output variable:
X = df[['weight','gender']]
y = df['likeness']
```

```python
X.head()
```

```python
y.head()
```

```python
#Machine learning algorithm
from sklearn.tree import DecisionTreeClassifier

#Create and fit our model
model  = DecisionTreeClassifier().fit(X,y)

# prediction
model.predict([[23,0]]) #  age of 23 and 0 defines the female it will predict what the person will like if she is 23 years old
```

```python
# How to measure the accuracy of our model
## split the data into test and train (80/20)

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test ,y_train, y_test = train_test_split(X,y, test_size=0.2) #80% training data 20% test data

#create a model
model = DecisionTreeClassifier().fit(X_train,y_train)

predicted_values = model.predict(X_test)
predicted_values
```

```python
#checking score
score= accuracy_score(y_test, predicted_values)
score
```

```python
# How to train and save your model.

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import joblib

model = DecisionTreeClassifier().fit(X,y)

joblib.dump(model , "foodie.joblib")
```

```python
#Visualization
from sklearn import tree
tree.export_graphviz(model,
                    out_file="foodie.dot",
                    feature_names=['age','gender'],
                    class_names=sorted(y.unique()),
                    label='all',
                    rounded=True,
                    filled=True)
```
